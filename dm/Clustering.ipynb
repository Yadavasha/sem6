```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.cluster import hierarchical
from sklearn.cluster import DBSCAN
import pandas as pd
```

```python
iris = pd.read_csv("iris.csv")
iris = iris.drop('Id',1)
```

```python
sep_l = iris.values[:,1]
pet_l = iris.values[:,2]
```

```python
for i in range(150):
    if i<=49:
        plt.plot(iris.values[i:,1],iris.values[i:,2],'ro')
    if i>49 and i<=99:
        plt.plot(iris.values[i:,1],iris.values[i:,2],'bo')
    if i>99:
        plt.plot(iris.values[i:,1],iris.values[i:,2],'go')        
plt.show()  
```

```python
estimatorKMeans = KMeans(n_clusters=3)
estimatorKMeans.fit(iris.values[:,1:3])
```

```python
for i in range(150):
    if estimatorKMeans.labels_[i]==0:
        plt.plot(iris.values[i:,1],iris.values[i:,2],'go')
    elif estimatorKMeans.labels_[i]==1:
        plt.plot(iris.values[i:,1],iris.values[i:,2],'ro')
    elif estimatorKMeans.labels_[i]==2:
        plt.plot(iris.values[i:,1],iris.values[i:,2],'bo')
plt.show()   
```

```python
estimateHierarchical=hierarchical.AgglomerativeClustering(n_clusters=3)
```

```python
estimateHierarchical.fit(iris.values[:,1:3])
```

```python
for i in range(150):
    if estimateHierarchical.labels_[i]==0:
        plt.plot(iris.values[i:,1],iris.values[i:,2],'go')
    elif estimateHierarchical.labels_[i]==1:
        plt.plot(iris.values[i:,1],iris.values[i:,2],'ro')
    elif estimateHierarchical.labels_[i]==2:
        plt.plot(iris.values[i:,1],iris.values[i:,2],'bo')
plt.show()        
```

```python
estimaterDbScan = DBSCAN(0.75)
estimatorDbScan.fit(iris.values[:,1:3])
```

```python
for i in range(150):
    if estimatorDbScan.labels_[i]==0:
        plt.plot(iris.values[i:,1],iris.values[i:,2],'go')
    elif estimatorDbScan.labels_[i]==1:
        plt.plot(iris.values[i:,1],iris.values[i:,2],'ro')
    elif estimatorDbScan.labels_[i]==2:
        plt.plot(iris.values[i:,1],iris.values[i:,2],'bo')
plt.show()       
```
